```python
import pandas as pd
from pathlib import Path
```

```python
pd.plotting.register_matplotlib_converters()
```

```python
datasets_dir = Path().absolute().parent.joinpath("datasets")
```

## Flight Delays (US Department of Transportation)

```python
flight_delays_df = pd.read_csv(f"{datasets_dir}/flight_delays.csv", index_col='Month')
```

```python
flight_delays_df.head()
```

### Bar Chart Plotting

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 4), dpi=96)

sns.barplot(
    y=flight_delays_df['NK'],
    x=flight_delays_df.index,
)

plt.title("NK Airlines - Avg Arrival Delays over Time")
plt.ylabel("Average Arrival Delays (mins)")
plt.xlabel("Month")
```

### Heatmap Plotting

```python
plt.figure(figsize=(20, 6))

sns.heatmap(
    data=flight_delays_df, 
    annot=True,
)

plt.title("Average Arrival Delay (mins) over Time")
plt.xlabel("Airlines")
plt.ylabel("Month")
```

## IGN Scores

```python
ign_df = pd.read_csv(f"{datasets_dir}/ign_scores.csv", index_col='Platform')
```

```python
ign_df.head()
```

```python
ign_df.loc['PC'].sort_values(ascending=False)
```

```python
ign_df.loc['PlayStation Vita'].sort_values(ascending=False)
```

### Bar chart Plotting

```python
plt.figure(figsize=(35, 5))

sns.barplot(
    y=ign_df['Racing'],
    x=ign_df.index, 
)

plt.title("Average Score of Racing Games per Platform")
plt.ylabel("Average Score")
plt.xlabel("Platform")
```

### Heatmap Ploting

```python
plt.figure(figsize=(20, 8))

sns.heatmap(
    data=ign_df, 
    annot=True,
)

plt.xlabel("Genre")
plt.ylabel("Platform")
plt.title("IGN Avg Rating per Genre per Platform")
```

```python
import pandas as pd
from pathlib import Path
```

```python
pd.plotting.register_matplotlib_converters()
```

```python
datasets_dir = Path().absolute().parent.joinpath("datasets")
```

## Spotify Dataset

```python
spotify_df = pd.read_csv(f"{datasets_dir}/spotify.csv", index_col='Date', parse_dates=True)
```

```python
spotify_df.head()
```

### Line chart Plotting

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 5))

sns.set_style("whitegrid")

sns.lineplot(
    data=spotify_df,
    y="Shape of You",
    x="Date",
    label="Shape Of You",

)

sns.lineplot(
    data=spotify_df,
    y="Despacito",
    x="Date", 
    label="Despacito",
)

plt.title("Shape of You stream count")
plt.legend()
plt.ylabel("Song Count")
plt.xlabel("Date")
```

## Museum Visitors Dataset

```python
museum_df = pd.read_csv(f"{datasets_dir}/museum_visitors.csv", index_col='Date', parse_dates=True)
```

```python
museum_df.head()
```

```python
""" Visits in July """
museum_df.loc['2018-07-01']
```

```python
""" Visits in October """
museum_df.loc['2018-10-01']
```

### Line chart Plotting

```python
plt.figure(figsize=(16, 4))

sns.set_style("whitegrid")

sns.lineplot(data=museum_df)

plt.title("LA Museum Visits over Time")
plt.legend()
plt.ylabel("Average Visitors Count")
plt.xlabel("Date")
```

```python
plt.figure(figsize=(16, 4))

sns.set_style("whitegrid")

sns.lineplot(
    data=museum_df,
    y="Avila Adobe",
    x=museum_df.index,
    label='Avila Adobe Museum',
)

plt.title("Avila Adobe Museum - Average Visitor Count")
plt.legend()
plt.ylabel("Average Visitor Count")
plt.xlabel("Date")
```

```python
import pandas as pd
from pathlib import Path
```

```python
pd.plotting.register_matplotlib_converters()
```

```python
datasets_dir = Path().absolute().parent.joinpath("datasets")
```

## Iris Dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = pd.read_csv(f"{datasets_dir}/iris.csv", index_col='Id')
```

```python
iris_df.head()
```

### Histogram Plot

```python
sns.histplot(
    data=iris_df['Petal Length (cm)'],
    fill=True,
)

plt.ylabel("Count")
plt.xlabel("Petal Length (cm)")
```

### KDE Plot

```python
sns.kdeplot(
    data=iris_df['Petal Length (cm)'],
    fill=True,
)

plt.ylabel("Count")
plt.xlabel("Petal Length (cm)")
```

### 2D KDE Plot

```python
sns.jointplot(
    x=iris_df['Petal Length (cm)'],
    y=iris_df['Sepal Length (cm)'],
    kind="kde",
)
```

## Color-coded Iris Datasets

```python
iris_set_df = pd.read_csv(f"{datasets_dir}/iris_setosa.csv", index_col='Id')
iris_ver_df = pd.read_csv(f"{datasets_dir}/iris_versicolor.csv", index_col='Id')
iris_vir_df = pd.read_csv(f"{datasets_dir}/iris_virginica.csv", index_col='Id')
```

```python
iris_set_df.head()
```

### Distribution Plot

```python
sns.displot(
    data=iris_df,
    x="Petal Length (cm)",
    label="Iris Setosa",

)
sns.displot(
    data=iris_ver_df,
    x="Petal Length (cm)",
    label="Iris Versicolor",
    kde=False,
)

sns.displot(
    data=iris_vir_df,
    x="Petal Length (cm)",
    label="Iris Virginica", 
    kde=False,
)

plt.ylabel("Count")
plt.xlabel("Petal Length (cm)")
```

### Histogram Plot

```python
sns.histplot(
    data=iris_set_df,
    x="Petal Length (cm)",
    label="Iris Setosa",
    color='blue',
)

sns.histplot(
    data=iris_ver_df,
    x="Petal Length (cm)",
    label="Iris Versicolor",
    color="orange",
)

sns.histplot(
    data=iris_vir_df,
    x="Petal Length (cm)",
    label="Iris Virginica",
    color="green",
)

plt.legend()
plt.xlabel("Petal Length (cm)")
plt.ylabel("Count")
```

### KDE Plot

```python
sns.kdeplot(
    data=iris_set_df,
    x="Petal Length (cm)",
    label="Iris Setosa",
    color="cyan",
    fill=True,
)

sns.kdeplot(
    data=iris_ver_df,
    x="Petal Length (cm)",
    label="Iris Versicolor",
    color="orange",
    fill=True,
)

sns.kdeplot(
    data=iris_vir_df,
    x="Petal Length (cm)",
    label="Iris Virginica",
    color="green",
    fill=True,
)

plt.legend()
plt.ylabel("Count")
plt.xlabel("Petal Length (cm)")
```

## Cancer Datasets

```python
benign_cancer_df = pd.read_csv(f"{datasets_dir}/cancer_b.csv", index_col='Id')
malignant_cancer_df = pd.read_csv(f"{datasets_dir}/cancer_m.csv", index_col='Id')
```

```python
benign_cancer_df.head()
```

```python
malignant_cancer_df.head()
```

```python
malignant_cancer_df.loc[842517]['Radius (mean)']
```

### Histogram Plot

```python
sns.histplot(
    data=benign_cancer_df,
    x="Area (mean)",
    label="Benign",
    color="blue",
)

sns.histplot(
    data=malignant_cancer_df,
    x="Area (mean)",
    label="Malignant",
    color="red",
)

plt.title("Cancer Mean Area Dist")
plt.legend()
plt.ylabel("Count")
plt.xlabel("Mean Area")
```

#### Conclusion
According to this distribution, malignant Cancers have a wider mean area.  
And thus, we can conclude that the higher cancerous area, the higher are the chances of it being Malignant 

### KDE Plot

```python
sns.kdeplot(
    data=benign_cancer_df,
    x="Radius (worst)",
    label="Benign Cancer",
    color="blue",
    fill=True,
)

sns.kdeplot(
    data=malignant_cancer_df,
    x="Radius (worst)",
    label="Malignant Cancer",
    color="red",
    fill=True,
)

plt.title("Cancer Mean Area Dist")
plt.legend()
plt.ylabel("Count")
plt.xlabel("Radius (worst)")
```

#### Conclusion
According to this distribution, a malignant cancer with a radius of '25' is most likely t 
to be classified as malignant

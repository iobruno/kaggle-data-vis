## Project Setup

```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
from pathlib import Path
```

```python
pd.plotting.register_matplotlib_converters()
```

```python
datasets_dir = str(Path(os.path.abspath('')).parent.joinpath("datasets"))
```

## Museum Visitors Dataset

```python
museum_df = pd.read_csv(f"{datasets_dir}/museum_visitors.csv", 
                        index_col='Date', parse_dates=True)
```

```python
museum_df.head()
```

### Line chart Plotting

```python
plt.figure(figsize=(20, 4))
sns.lineplot(data=museum_df)

plt.title("LA Museum Visits over Time")
plt.xlabel("Date")
plt.ylabel("Average Visitors Count")
```

```python
plt.figure(figsize=(20, 4))
sns.lineplot(data=museum_df['Avila Adobe'])

plt.title("Avila Adobe Museum - Average Visitor Count")
plt.xlabel("Date")
plt.ylabel("Average Visitor Count")
```

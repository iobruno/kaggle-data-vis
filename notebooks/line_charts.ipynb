## Project Setup

```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
from pathlib import Path
```

```python
pd.plotting.register_matplotlib_converters()
```

```python
datasets_dir = str(Path(os.path.abspath('')).parent.joinpath("datasets"))
```

## Spotify Dataset

```python
spotify_df = pd.read_csv(f"{datasets_dir}/spotify.csv", index_col='Date', parse_dates=True)
```

```python
spotify_df.head()
```

### Line chart Plotting

```python
plt.figure(figsize=(12, 5))
sns.set_style("whitegrid")

sns.lineplot(x='Date', y='Shape of You',
             label='Shape Of You',
             data=spotify_df)

sns.lineplot(x='Date', y='Despacito',
             label='Despacito',
             data=spotify_df)

plt.xlabel("Date")
plt.ylabel("Song Count")
plt.title("Shape of You stream count")
plt.legend()
```

## Museum Visitors Dataset

```python
museum_df = pd.read_csv(f"{datasets_dir}/museum_visitors.csv", 
                        index_col='Date', parse_dates=True)
```

```python
museum_df.head()
```

```python
""" Visits in July """
museum_df.loc['2018-07-01']
```

```python
""" Visits in October """
museum_df.loc['2018-10-01']
```

### Line chart Plotting

```python
plt.figure(figsize=(16, 4))
sns.set_style("whitegrid")

sns.lineplot(data=museum_df)

plt.title("LA Museum Visits over Time")
plt.xlabel("Date")
plt.ylabel("Average Visitors Count")
```

```python
plt.figure(figsize=(16, 4))
sns.set_style("whitegrid")

sns.lineplot(x=museum_df.index,
             y='Avila Adobe',
             label='Avila Adobe Museum',
             data=museum_df)

plt.title("Avila Adobe Museum - Average Visitor Count")
plt.xlabel("Date")
plt.ylabel("Average Visitor Count")
plt.legend()
```
